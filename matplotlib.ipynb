```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = [1,2,3,4]
b = [5,6,7,8]
```

```python
plt.figure(figsize = (2,2))

plt.plot(a,b, )
plt.xlabel("this is a x axis ")
plt.ylabel("This is a y axis")
plt.title("This is title of the graph")
```

```python
x = np.random.normal(0,2,50)
y = np.random.normal(0,2, len(x))
plt.figure(figsize=(5,4))

plt.scatter(x,y, s = 200, marker = "v", c= "black", alpha= 0.4)
plt.xlabel("This is an x axis statement")
plt.ylabel("This is a y axis statemnt")
plt.title("This is the title of the graph")
```

```python
# subplot in matplotlib is used to create multiple plots(visualization) using same data sets
```

```python
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])



plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("line plot 1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("this is second line ")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("this is third plot")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("this is fourth plot")




plt.tight_layout()
plt.show()
```

```python

# question 5
a box plot also know as box whisker plot graphical representation of data set
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)

plt.title('Box Plot Example')
plt.xlabel('Data')
plt.ylabel('Values')

plt.show()
```
